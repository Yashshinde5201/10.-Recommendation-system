# import Libraries

```python
import pandas as pd
import numpy as np
```

# read dataset

```python
books = pd.read_csv('book.csv',encoding='latin-1')
books.head()
```

# droping unwanted column

```python
books.drop('Unnamed: 0',inplace=True,axis=1)
books.head()
```

# renaming column names

```python
books=books.rename(columns={'User.ID':'ID', 'Book.Title':'Title', 'Book.Rating':'Rating'})
books.head()
```

# sorting id column from asending order

```python
books=books.sort_values('ID')
books.head(10)
```

# deleting duplicated entries

```python
books=books.drop_duplicates()
```

# duplicate entry of user with same book

```python
books[books.duplicated(subset=["ID","Title"])]
```

```python
bookclean=books.drop_duplicates(subset=["ID","Title"])
bookclean
```

# sparse matrix

```python
user_books = bookclean.pivot(index='ID',
                            columns='Title',
                            values='Rating')
```

```python
user_books.head()
```

```python
#impute those NaN with 0 values
user_books.fillna(0, inplace=True)
```

```python
user_books
```

# Calculating Cosine Similarity between Users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books.values, metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#set the index and column names to user to user ids because above it automattically assign index no. not a user id 
user_sim_df.index = books.ID.unique()
user_sim_df.columns = books.ID.unique()
```
